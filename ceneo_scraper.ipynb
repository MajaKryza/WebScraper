### Analiza struktury pojedynczej opinii

##

|składowa|nazwa|selektor|
|--------|-----|--------|
|opinia|opinion|div.js_product-review|
|identyfikator|opinion_id|div.js_product-review[data-entry-id]|
|autor|author|span.user-post__author-name |
|treść|content|div.user-post__text|
|ocena|score|span.user-post__score-count|
|rekomendacja|recommendation|span.user-post__author-recomendation > em|
|listaa zalet|pros|div.review-feature__item--positive|
|lista wad|cons|div.review-feature__item--negative|
|dla ilu przydatna|thumbs_up|button.vote-yes > span|
|dla ilu nieprzydatna|thumbs_down|button.vote-no > span|
|dat zamieszczenia|post_date|span.user-post__published > time:nth-child(1)[datetime]|
|data zakupu|purchase_date|span.user-post__published > time:nth-child(2)[datetime]|

```python
from bs4 import BeautifulSoup

def parse_ceneo_reviews(html_content):
    soup = BeautifulSoup(html_content, 'html.parser')
    reviews = []

    # Znalezienie wszystkich kontenerów opinii
    review_elements = soup.select('div.js_product-review')

    for element in review_elements:
        # Podstawowe dane
        review_id = element.get('data-entry-id')
        author = element.select_one('span.user-post__author-name').get_text(strip=True)
        
        # Ocena i rekomendacja
        score = element.select_one('span.user-post__score-count').get_text(strip=True)
        recommendation = element.select_one('em.recommended')
        is_recommended = recommendation.get_text(strip=True) if recommendation else "Brak"
        
        # Tekst opinii
        content = element.select_one('div.user-post__text').get_text(strip=True)
        
        # Daty (wystawienia i zakupu)
        times = element.select('span.user-post__published time')
        date_published = times[0]['datetime'] if len(times) > 0 else None
        date_bought = times[1]['datetime'] if len(times) > 1 else None

        # Zalety i Wady (lista)
        pros = [span.get_text(strip=True) for span in element.select('div.review-feature__item--positive')]
        cons = [span.get_text(strip=True) for span in element.select('div.review-feature__item--negative')]

        # Głosy przydatności
        useful_yes = element.select_one('button.vote-yes span').get_text(strip=True)
        useful_no = element.select_one('button.vote-no span').get_text(strip=True)

        reviews.append({
            "review_id": review_id,
            "author": author,
            "score": score,
            "recommendation": is_recommended,
            "content": content,
            "date_published": date_published,
            "date_bought": date_bought,
            "pros": pros,
            "cons": cons,
            "useful_yes": useful_yes,
            "useful_no": useful_no
        })

    return reviews

# Przykład użycia:
html_data = "https://www.ceneo.pl/124893467#tab=reviews"
result = parse_ceneo_reviews(html_data)
print(result[0])
```
